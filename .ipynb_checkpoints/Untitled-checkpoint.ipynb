```python
from wallFlux import qw as qw1
from wallFlux import tauw as tauw1

from wallFlux_2 import qw as qw2
from wallFlux_2 import tauw as tauw2
import pandas as pd

import numpy as np

from numDeri import numericalDerivative as deri
```

```python
MainDir = '../../WM_ZPGBL/Data/Stats/'

WM = {} # initiate to fill later

WM['nx'] = 16385
WM['ny'] = 501 
WM['ntime'] = 37
WM['nu'] = 1/80
WM['Pr'] = 1
WM['rho'] = 1
WM['cp'] = 1 # double check
WM['mu'] = WM['rho']*WM['nu']
WM['alpha'] = WM['nu']/WM['Pr']
WM['Uinf'] = 1
WM['Tw'] = 1
WM['Tinf'] = 0

temp = pd.read_table(MainDir + 'i_and_x_versus_Re_theta_and_Re_tau.dat', header=None, delimiter=' ');
temp = np.array(temp)
WM['x'] = temp[:, 2]
WM['Retheta'] = temp[:,3]
WM['Retau'] = temp[:,4]

temp = pd.read_table(MainDir + 'Re_theta_versus_delta.dat', header=None, delimiter=' ');
temp = np.array(temp)
WM['delta'] = temp[:,2]

temp = pd.read_table(MainDir + 'Re_theta_versus_cf.dat', header=None, delimiter=' ');
temp = np.array(temp)
WM['Cf'] = temp[:,2]
WM['tauw'] = 0.5*WM['rho']*WM['Uinf']**2*WM['Cf']

temp = pd.read_table(MainDir + 'Re_theta_versus_utau.dat', header=None, delimiter=' ');
temp = np.array(temp)
WM['utau'] = temp[:,2]

# read stats for sample Retheta

WM['RethetaSample'] = [100, 125, 150, 175, 200, 300, 375, 670,
                 900, 1000, 1200, 1410, 1750, 2000, 2540, 2900]
WM['nSample'] = len(WM['RethetaSample'])

WM['RetauSample'] = np.zeros((1, WM['nSample']))

WM['deltaSample'] = np.zeros((1, WM['nSample'])) # BL thickness
WM['deltastarSample'] = np.zeros((1, WM['nSample'])) # displacemnet thickness
WM['HSample'] = np.zeros((1, WM['nSample'])) # Shape factor
WM['thetaSample'] = np.zeros((1, WM['nSample'])) # momentum thickness

WM['CfSample'] = np.zeros((1, WM['nSample']))
WM['tauwSample'] = np.zeros((1, WM['nSample']))
WM['utauSample'] = np.zeros((1, WM['nSample']))

WM['Uplus']= np.zeros((401, WM['nSample']))
WM['yplus']= np.zeros((401, WM['nSample']))
WM['dUplusdyplus'] = np.zeros((401, WM['nSample']))
WM['uvplus']= np.zeros((498, WM['nSample']))
WM['yplus_uv']= np.zeros((498, WM['nSample']))
WM['uuplus']= np.zeros((501, WM['nSample']))
WM['yplus_uu']= np.zeros((501, WM['nSample']))

for ii in range(WM['nSample']):
  #  find the closest Retheta to RethetaSample
    ind = np.argmin(np.abs(WM['Retheta'] - WM['RethetaSample'][ii]))
    WM['RetauSample'][0, ii] = WM['Retau'][ind]
    WM['CfSample'][0, ii] = WM['Cf'][ind]
    WM['tauwSample'][0, ii] = WM['tauw'][ind]
    WM['utauSample'][0, ii] = WM['utau'][ind]
    
    temp = pd.read_table(MainDir + 'Re_theta_versus_delta.dat', header=None, delimiter=' ');
    temp = np.array(temp)
    WM['deltaSample'][0, ii] = temp[ind, 2]

    temp = pd.read_table(MainDir + 'Re_theta_versus_delta_star.dat', header=None, delimiter=' ');
    temp = np.array(temp)
    WM['deltastarSample'][0, ii] = temp[ind, 2]

    temp = pd.read_table(MainDir + 'Re_theta_versus_shape_factor.dat', header=None, delimiter=' ');
    temp = np.array(temp)
    WM['HSample'][0, ii] = temp[ind, 2]
    
    temp = pd.read_table(MainDir + 'yplus_versus_U_plus_at_Re_theta_' + str(WM['RethetaSample'][ii]).zfill(4) + '.dat',
                         header=None, delimiter=' ');
    temp = np.array(temp)
    WM['yplus'][:, ii] = temp[:, 1]
    WM['Uplus'][:, ii] = temp[:, 2]
    WM['dUplusdyplus'][:, ii] = deri(temp[:, 1], temp[:, 2])
    

    temp = pd.read_table(MainDir + 'yplus_versus_uv_plus_at_Re_theta_' + str(WM['RethetaSample'][ii]).zfill(4) + '.dat',
                         header=None, delimiter=' ');
    temp = np.array(temp)
    WM['yplus_uv'][:, ii] = temp[:, 1]
    WM['uvplus'][:, ii] = temp[:, 2]
    
    temp = pd.read_table(MainDir + 'yplus_versus_urms_plus_at_Re_theta_' + str(WM['RethetaSample'][ii]).zfill(4) + '.dat',
                         header=None, delimiter=' ');
    temp = np.array(temp)
    WM['yplus_uu'][:, ii] = temp[:, 1]
    WM['uuplus'][:, ii] = temp[:, 2]**2

```

```python
import matplotlib.pyplot as plt

ii = 10
nu = WM['nu']
rho = WM['rho']
utau = WM['utau'][ii]

Up = WM['Uplus'][:10, ii]
U = Up*utau
yp = WM['yplus'][:10, ii]
y = yp*nu/utau
uvp = WM['uvplus'][:10, ii]
uv = uvp*utau**2

t1, te1 = qw1(y, U, uv, 1, nu, rho)
t2, te2 = qw2(y, U, uv, 1, nu, rho)

print(rho*utau**2, t1, t2)
```
